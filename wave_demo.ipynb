```python
import pywt
import numpy as np
import pandas as pd
from plotnine import *
```

```python
data1 = np.loadtxt(r"C:\Users\UPC\Desktop\demo.txt")

wavelet = 'db1'
level = 1
# 
df = pd.DataFrame({
    'col1': np.arange(len(data1)),
    'col2': data1
})
```

```python

print(df.shape)
date = np.arange(0,len(data1)-1)
coeffs = pywt.wavedec(data1, wavelet, level=level,mode = 'periodic')
wave_data = pd.DataFrame({
    'wave1': coeffs[0],
    'wave2': coeffs[1],
})
wave_data.to_csv('wave_data.csv',index = False)
ggplot(df, aes(x='col1', y='col2')) + geom_line(color='blue') + theme(axis_text=element_blank(),  axis_title=element_blank(), axis_ticks=element_blank(), panel_background = element_rect(fill='white'), figure_size=(1, 0.4))
plot = (ggplot(df, aes(x='col1', y='col2')) + geom_line(color='blue') + theme(axis_text=element_blank(),  axis_title=element_blank(), axis_ticks=element_blank(), panel_background = element_rect(fill='white'), figure_size=(1, 0.4)) ) # 去掉刻度线
plot.save("origin.tif",dpi=600)
```

```python

waveall = pywt.waverec(coeffs, wavelet)
wave_all = pd.DataFrame({
    'reshape': waveall,
    'rigin': data1,
})
wave_all.to_csv('wave_reshape.csv',index = False)
```
